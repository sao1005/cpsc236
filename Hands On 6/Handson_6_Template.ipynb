## 12.1 Champion Counter
Create a program that reads a text file that contains a list of FIFA World Cup champions and determines the country that has won the most championships.

### Console:
```powershell
Country	        Wins	Years	
=======	        ====	=====	
Argentina       2       1978, 1986
Brazil	        5       1958, 1962, 1970, 1994, 2002
England	        1       1966		
France	        1       1998		
Germany	        4       1954, 1974, 1990, 2014
Italy	        4       1934, 1938, 1982, 2006
Spain	        1       2010
Uruguay	        2       1930, 1950
```

### Specifications:
- Use the attached text file named `world_cup_champions.txt` that contains data like this: 
```powershell
Year,Country,Coach,Captain 
1930,Uruguay,Alberto Suppici,Jos√© Nasazzi 
1934,Italy,Vittorio Pozzo,Gianpiero Combi 
1938,Italy,Vittorio Pozzo,Giuseppe Meazza
...
```
- When the program starts, it should read the text file and use a dictionary to store the required data using the name of each country that has won the World Cup as the key.
- The program should compile the data shown above and display the countries alphabetically.



````python
import csv
def main():
    champions,winnerList=wordCount()
    
    display(champions=champions,winnerList=winnerList)
    
    
def wordCount():
    winnerList={}
    with open('world_cup_champions.txt', 'r', encoding='utf-8-sig') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            if row[1] in winnerList:
                winnerList[row[1]][0]+=1
                winnerList[row[1]][1].append(row[0])
            else:
                winnerList[row[1]]=[1,[row[0]]]
    
    champions=sorted(winnerList.keys())
    
    return champions,winnerList

def display(champions,winnerList):
    print("Country	        Wins	Years	\n=======	        ====	=====")
    for i in range(len(champions)):
        if len(champions[i])<7:
            tabs='\t\t\t'
        else:
            tabs='\t\t'
        print(champions[i],tabs,winnerList[champions[i]][0],'\t\t',', '.join(winnerList[champions[i]][1]))
    
    

if __name__=="__main__":
    main()
````

## 12.2 Monthly Sales
Create a program that allows you to view and edit the sales amounts for each month of the current year.

### Console:
```powershell

edit	- Edit sales for specified month

Command: edit

Sales Amount: 15293

Command: totals
Yearly total: 67,855.00
Monthly average: 0

Command: view
Three-letter Month: july 
Invalid three-letter month.

Command: exit Bye!
```

### Specifications:
- Use the attached text file named `monthly_sales.txt` that consists of rows that contain three-letter abbreviations for the month and the monthly sales.
- The program should read the file and store the sales data for each month in a dictionary with the month abbreviation as the key for each item.
- Whenever the sales data is edited, the program should write the changed data to the text file.




````python
import sys

def main():
    monthly={}
    with open("monthly_sales.txt",'r') as file:
        for line in file:
            month,amount=line.split()
            amount=float(amount)
            monthly.update({month: amount})
    
    
    
    print("Monthly Sales program")
    print("COMMAND MENU\nview	- View sales for specified month\nedit	- Edit sales for specified month\ntotals - View sales summary for year\nexit	- Exit program")
    correct=["view","edit","totals","exit"]
    while True:
        choice=input()
        while choice not in correct:
            choice=input("Invalid input, try again: ")
        if choice.lower()=="view":
            view(monthly=monthly)
        elif choice.lower()=="edit":
            edit(monthly=monthly)
        elif choice.lower()=="totals":
            total(monthly=monthly)
        else:
            sys.exit()
    


def view(monthly):
    print("Command View")
    choice=input("Three-letter Month: ")
    while choice.capitalize() not in monthly:
        choice=input("Invalid three-letter month. Try again: ")
    print("${:,.2f}".format(monthly[choice.capitalize()]))
    
def edit(monthly):
    print("Command edit")
    choice=input("Three-letter Month: ")
    while choice.capitalize() not in monthly:
        choice=input("Invalid three-letter month. Try again: ")
    change=float(input("Change "+choice.capitalize()+" sales amount to: "))
    monthly[choice.capitalize()]=change
    
    with open("monthly_sales.txt","w") as file:
        for key,value in monthly.items():
            file.write('%s %s\n' % (key, value))
        
    #add changed list to file

def total(monthly):
    total=0.0
    for month in monthly:
        total=total+monthly[month]
    
    print("Yearly total is",total)
    print("Monthly average is ", total/12)
        


if __name__=="__main__":
    main()
````
